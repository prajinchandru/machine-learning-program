```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
saldf= pd.read_csv("/content/drive/MyDrive/Salary Data.csv")
```

```python
saldf.head()
```

```python
saldf.info()
```

```python
saldf.isnull().sum()
```

```python
inp = saldf[['Years of Experience']]
out = saldf['Salary']
```

```python
LR = LinearRegression()
```

```python
train_data = pd.concat([inp,out],axis=1)
train=train_data.dropna()
```

```python
inp1=train[['Years of Experience']]
out1=train['Salary']
```

```python
LR.fit(inp1,out1)
```

```python
LR.predict([[5]])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Apply Label Encoding to 'Gender'
saldf['Gender_Encoded'] = le.fit_transform(saldf['Gender'])

# Apply Label Encoding to 'Education Level'
saldf['Education_Level_Encoded'] = le.fit_transform(saldf['Education Level'])

# Apply Label Encoding to 'Job_Title'
saldf['Job_Title_Encoded'] = le.fit_transform(saldf['Job Title'])


print("Original 'Gender' values and their encoded counterparts:")
print(saldf[['Gender', 'Gender_Encoded']].drop_duplicates().sort_values(by='Gender_Encoded'))

print("\nOriginal 'Education Level' values and their encoded counterparts:")
print(saldf[['Education Level', 'Education_Level_Encoded']].drop_duplicates().sort_values(by='Education_Level_Encoded'))

print("\nOriginal 'Job Title' values and their encoded counterparts:")
print(saldf[['Job Title', 'Job_Title_Encoded']].drop_duplicates().sort_values(by='Job_Title_Encoded'))


display(saldf.head())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(inp1, out1, color='blue', label='Actual Data')
plt.plot(inp1, LR.predict(inp1), color='red', label='Regression Line')
plt.title('Years of Experience vs Salary (Linear Regression)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
